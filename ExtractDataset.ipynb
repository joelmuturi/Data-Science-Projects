```python
import pandas as pd

# Load the dataset
file_path = '/mnt/data/1. Weather Data.csv'
data = pd.read_csv('C:/Users/Joel/kaggle_datasets/weather_dataset/1. Weather Data.csv')

# 1. Find all records where the weather was exactly clear.
clear_weather_records = data[data['Weather'] == 'Clear']
print("Clear Weather Records:\n", clear_weather_records.head(), "\n")

# 2. Find the number of times the wind speed was exactly 4 km/hr.
wind_speed_4_count = data[data['Wind Speed_km/h'] == 4].shape[0]
print("Wind Speed 4 km/hr Count:", wind_speed_4_count, "\n")

# 3. Check if there are any NULL values present in the dataset.
null_values = data.isnull().sum().sum()
print("Number of NULL values:", null_values, "\n")

# 4. Rename the column "Weather" to "Weather_Condition."
data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
print("Columns after renaming:\n", data.columns, "\n")

# 5. Calculate the mean visibility of the dataset.
mean_visibility = data['Visibility_km'].mean()
print("Mean Visibility:", mean_visibility, "\n")

# 6. Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
wind_speed_greater_24_vis_25 = data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].shape[0]
print("Records with wind speed > 24 km/hr and visibility = 25 km:", wind_speed_greater_24_vis_25, "\n")

# 7. Calculate the mean value of each column for each weather condition.
mean_values_per_weather = data.groupby('Weather_Condition').mean(numeric_only=True)
print("Mean values per weather condition:\n", mean_values_per_weather, "\n")

# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
clear_humidity_vis = data[(data['Weather_Condition'] == 'Clear') & ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))]
print("Clear weather with humidity > 50 or visibility > 40:\n", clear_humidity_vis.head(), "\n")

# 9. Find the number of weather conditions that include snow.
snow_conditions = data[data['Weather_Condition'].str.contains('Snow', case=False)].shape[0]
print("Number of weather conditions that include snow:", snow_conditions, "\n")
```
